```python
df_raw = pd.read_csv(filename)
df_raw = df_raw.drop('Unnamed: 0',axis=1)
df_raw
```

```python
df_sample = df_raw.sample(150000)
df_sample
```

```python
df_sample.isnull().sum()
```

```python
X_train = np.array(df_sample.iloc[:100000, [9,11,13]])
X_test = np.array(df_sample.iloc[100000:, [9,11,13]])
Y_train = np.array(df_sample.iloc[:100000, 8])
Y_train = Y_train.ravel()
Y_test = np.array(df_sample.iloc[100000:, 8])
Y_test = Y_test.ravel()
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```
